```python
try:
    from datasets import load_dataset
except ImportError:
    !pip install datasets
    from datasets import load_dataset
imdb = load_dataset("imdb")
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification

checkpoint = "distilbert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(checkpoint)
model = AutoModelForSequenceClassification.from_pretrained(checkpoint, num_labels=2)
```

```python
def tokenize(inp):
    return tokenizer(inp["text"], truncation=True)

tokenized_imdb = imdb.map(tokenize, batched = True)
```

```python
tokenized_imdb = tokenized_imdb.remove_columns(["text"])
tokenized_imdb.set_format("torch")
```

```python
tokenized_imdb["train"][0]
```

```python
from transformers import DataCollatorWithPadding

data_collator = DataCollatorWithPadding(tokenizer=tokenizer)
```

```python
import numpy as np
from sklearn.metrics import accuracy_score

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)
    return {"accuracy": accuracy_score(labels, predictions)}
```

```python
from transformers import TrainingArguments, Trainer

training_args = TrainingArguments(
    output_dir = "./results",
    eval_strategy = "epoch",
    save_strategy = "epoch",
    learning_rate = 2e-5,
    per_device_train_batch_size = 64,
    per_device_eval_batch_size = 64,
    num_train_epochs = 3,
    weight_decay = 0.01,
    logging_dir  = "./logs",
    logging_steps = 10,
    push_to_hub = False
)
```

```python
from transformers import Trainer


trainer = Trainer(
    model = model,
    args = training_args,
    train_dataset = tokenized_imdb["train"],
    eval_dataset = tokenized_imdb["test"],
    tokenizer = tokenizer,
    data_collator = data_collator,
    compute_metrics = compute_metrics
)
```

```python
trainer.train()
```

```python

tokenizer.save_pretrained("sentiment_model")
```

```python
from transformers import pipeline

classifier = pipeline("text-classification", model="./model", tokenizer="sentiment_model")
```

```python
text = "I really love this movie, it was fantastic!"
print(classifier(text))
```

```python
label_map = {0: "NEGATIVE", 1: "POSITIVE"}

text = "I enjoy movie"
result = classifier(text)[0]  # Get first prediction

# Convert label
result['label'] = label_map[int(result['label'].split('_')[-1])]
print(result)
```
